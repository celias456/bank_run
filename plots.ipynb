```python
import pandas as pd
import matplotlib.pyplot as plt
```

Generate the threshold-type plot

```python
# Read in data from the csv files
one_scattered = pd.read_csv("1.csv")
one_clustered = pd.read_csv("6.csv")
heterogeneous_uniform = pd.read_csv("11.csv")
heterogeneous_normal = pd.read_csv("16.csv")

# Filter the data
one_scattered_filtered = one_scattered[(one_scattered.iteration == 0)]
one_clustered_filtered = one_clustered[(one_clustered.iteration == 0)]
heterogeneous_uniform_filtered = heterogeneous_uniform[(heterogeneous_uniform.iteration == 0)]
heterogeneous_normal_filtered = heterogeneous_normal[(heterogeneous_normal.iteration == 0)]

# Get variables from the data
one_scattered_month = one_scattered_filtered["Step"] + 1
one_scattered_runners_percent = one_scattered_filtered["adopters_percent"]
one_scattered_liquid_assets = one_scattered_filtered["Required_Liquidity"] / 1000
one_scattered_deposits = one_scattered_filtered["Deposits"]
one_scattered_data = {'Month': one_scattered_month, 'Percentage of Runners': one_scattered_runners_percent , 'Reserves': one_scattered_liquid_assets, 'Deposits': one_scattered_deposits}
one_scattered_df = pd.DataFrame(one_scattered_data)

one_clustered_month = one_clustered_filtered["Step"] + 1
one_clustered_runners_percent = one_clustered_filtered["adopters_percent"]
one_clustered_liquid_assets = one_clustered_filtered["Required_Liquidity"] / 1000
one_clustered_deposits = one_clustered_filtered["Deposits"]
one_clustered_data = {'Month': one_clustered_month, 'Percentage of Runners': one_clustered_runners_percent , 'Reserves': one_clustered_liquid_assets, 'Deposits': one_clustered_deposits}
one_clustered_df = pd.DataFrame(one_clustered_data)

heterogeneous_uniform_month = heterogeneous_uniform_filtered["Step"] + 1
heterogeneous_uniform_runners_percent = heterogeneous_uniform_filtered["adopters_percent"]
heterogeneous_uniform_liquid_assets = heterogeneous_uniform_filtered["Required_Liquidity"] / 1000
heterogeneous_uniform_deposits = heterogeneous_uniform_filtered["Deposits"]
heterogeneous_uniform_data = {'Month': heterogeneous_uniform_month, 'Percentage of Runners': heterogeneous_uniform_runners_percent , 'Reserves': heterogeneous_uniform_liquid_assets, 'Deposits': heterogeneous_uniform_deposits}
heterogeneous_uniform_df = pd.DataFrame(heterogeneous_uniform_data)

heterogeneous_normal_month = heterogeneous_normal_filtered["Step"] + 1
heterogeneous_normal_runners_percent = heterogeneous_normal_filtered["adopters_percent"]
heterogeneous_normal_liquid_assets = heterogeneous_normal_filtered["Required_Liquidity"] / 1000
heterogeneous_normal_deposits = heterogeneous_normal_filtered["Deposits"]
heterogeneous_normal_data = {'Month': heterogeneous_normal_month, 'Percentage of Runners': heterogeneous_normal_runners_percent , 'Reserves': heterogeneous_normal_liquid_assets, 'Deposits': heterogeneous_normal_deposits}
heterogeneous_normal_df = pd.DataFrame(heterogeneous_normal_data)

# Create the data frames array
dfs = [one_scattered_df, one_clustered_df, heterogeneous_uniform_df, heterogeneous_normal_df]

# Create titles for plots
titles = ['One-Scattered', 'One-Clustered', 'Heterogeneous Uniform', 'Heterogeneous Normal']

# Create 2x2 subplots
fig, axs = plt.subplots(2, 2, figsize=(14, 10), sharex=False)
axs = axs.flatten()  # Flatten to iterate easily

# Store handles for legend
legend_handles = []
legend_labels = []

for i, (df, ax, title) in enumerate(zip(dfs, axs, titles)):
    # Left Y-axis
    l1, = ax.plot(df['Month'], df['Deposits'], label='Deposits')
    l2, = ax.plot(df['Month'], df['Reserves'], label='Reserves', linestyle='--')
    ax.set_ylabel("Millions of Dollars")
    ax.set_title(title)

    # Right Y-axis
    ax2 = ax.twinx()
    l3, = ax2.plot(df['Month'], df['Percentage of Runners'], label='Percentage of Runners', linestyle=':')
    ax2.set_ylabel("Percentage")

    # Only add legend labels once (avoid duplicates)
    if i == 0:
        legend_handles = [l1, l2, l3]
        legend_labels = [h.get_label() for h in legend_handles]

# Add common legend outside the plots
fig.legend(legend_handles, legend_labels, loc='upper center', ncol=3, bbox_to_anchor=(0.5, 0))

# Adjust layout to fit legend
plt.tight_layout(rect=[0, 0, 1, 0.97])

# Add a super title
plt.suptitle('Threshold-Type')

plt.show()
```
